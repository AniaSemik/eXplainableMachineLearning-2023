## XAI Homework 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn import preprocessing, tree
from sklearn.metrics import explained_variance_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Loading data

Overview of the data

```python
data = pd.read_csv("Maths.csv")
print(data.head())

for c in data.columns:
  if type(data[c][0]) == str:
    le = preprocessing.LabelEncoder()
    le.fit(data[c])
    data[c] = le.transform(data[c])
```

Data summary

```python
for c in data.columns:
  print(c, data[c].unique())
```

dividing the data into training and testing batches

```python
X, y = data[data.columns[:-1]], data[data.columns[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=420)
```

### Model #1: logistic regression
training the logistic regression model

```python
lreg = LinearRegression().fit(X_train, y_train)
y_pred_lreg = lreg.predict(X_test)
```

visualisation of how the model performs

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
fig.suptitle("Logistic regression outcomes visualised according to grades from the previous semesters")
ax[0].plot(X_test.iloc[:,-2], y_test, ".", label="true values")
ax[0].plot(X_test.iloc[:,-2], y_pred_lreg, ".", label="predicted values")
ax[0].set_xlabel(X_test.columns[-2])
ax[0].set_ylabel("final grade")
ax[0].set_title("first semester")
ax[1].plot(X_test.iloc[:,-1], y_test, ".", label="true values")
ax[1].plot(X_test.iloc[:,-1], y_pred_lreg, ".", label="predicted values")
ax[1].set_xlabel(X_test.columns[-1])
ax[1].set_title("second semester")
ax[1].set_ylabel("final grade")
plt.legend()
```

Other than 0's, the true and predicted values seem to have a similar distribution. When it comes to 0's, predicted values sometimes have values below 0, which do not occur in the actual data.

### Model #2: boosting
learning the model

```python
boost = XGBRegressor()
boost.fit(X_train, y_train)
y_pred_boost = boost.predict(X_test)
```

visualisation of how the model performs

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
fig.suptitle("Boosting model outcomes visualised according to grades from the previous semesters")
ax[0].plot(X_test.iloc[:,-2], y_test, ".", label="true values")
ax[0].plot(X_test.iloc[:,-2], y_pred_boost, ".", label="predicted values")
ax[0].set_xlabel(X_test.columns[-2])
ax[0].set_ylabel("final grade")
ax[0].set_title("first semester")
ax[1].plot(X_test.iloc[:,-1], y_test, ".", label="true values")
ax[1].plot(X_test.iloc[:,-1], y_pred_boost, ".", label="predicted values")
ax[1].set_xlabel(X_test.columns[-1])
ax[1].set_title("second semester")
ax[1].set_ylabel("final grade")
plt.legend()
```

The predicted values for 2nd semester seem to more closely resemble the true values than for the 1st semester. There are only one occurences of values below 0 in both semesters, which is better than in the previous model.

### Model #3: Random forest
learning the random forest

```python
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

visualising one of the trees in the forest

```python
plt.figure(figsize=(20,8))
tree.plot_tree(rf.estimators_[18], filled=True, impurity=True, rounded=True);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
fig.suptitle("Random forest outcomes visualised according to grades from the previous semesters")
ax[0].plot(X_test.iloc[:,-2], y_test, ".", label="true values")
ax[0].plot(X_test.iloc[:,-2], y_pred_rf, ".", label="predicted values")
ax[0].set_xlabel(X_test.columns[-2])
ax[0].set_ylabel("final grade")
ax[0].set_title("first semester")
ax[1].plot(X_test.iloc[:,-1], y_test, ".", label="true values")
ax[1].plot(X_test.iloc[:,-1], y_pred_rf, ".", label="predicted values")
ax[1].set_xlabel(X_test.columns[-1])
ax[1].set_title("second semester")
ax[1].set_ylabel("final grade")
plt.legend()
```

No predicted values go below 0, however also no values reach 0 when they are supposed to. Values for the 2nd semester look too linear compared to the true values.

## Analyses
of how the models performed in comparison to each another

### Explained variance

```python
print("Linear regression: %0.2f\nBoosting: %0.2f\nRandom forest: %0.2f"
%tuple(explained_variance_score(y_test, y_pred) for y_pred in [y_pred_lreg, y_pred_boost, y_pred_rf]))
```

When it comes to the % of explained variance, all models performed similarily well, however linear regression has the lowest score.

### Histogram of values from each of the models against the true values from the test dataset

```python
plt.hist(y_test, alpha=1, label='original data')
plt.hist(y_pred_rf, alpha=0.4, label='Random forest')
plt.hist(y_pred_lreg, alpha=0.4, label='Linear regression')
plt.hist(y_pred_boost, alpha=0.4, label='Boosting')
plt.legend()
```

All 3 models have the same general shape of the original data, however the boosting model most closely resembles the original data around 10, and random forest around 0 (again, only this model does not reach values below 0).

### Mean absolute error for the classifiers

```python
print("Linear regression: %0.2f\nBoosting: %0.2f\nRandom forest: %0.2f"
%tuple(mean_absolute_error(y_test, y_pred) for y_pred in [y_pred_lreg, y_pred_boost, y_pred_rf]))
```

Random forest and boosting have a similar score, better than linear regression.

## Conclusions

Linear regression definitely performs worst out of the 3 models. Boosting and random forest have similar stats and the choice of one of those two models should rest on the user and what is expected of the model.
